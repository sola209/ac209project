```python
with open('reviews.json', 'rb') as f:
    data = f.readlines()
    
data = map(lambda x: x.rstrip(), data)

data_json_str = "[" + ','.join(data) + "]"

data_df = pd.read_json(data_json_str)
```

```python
years = []
months = []
days = []
for date in data_df['reviewTime']:
    splitted = date.split(' ')
    years.append(splitted[2])
    months.append(splitted[0])
    day = splitted[1]
    days.append(day[:-1])
    
date_df = pd.DataFrame({'year': years,
                       'month': months,
                       'day': days})

print date_df.iloc[0]
final_dates = pd.to_datetime(date_df)
```

```python
data_df['datetimes'] = final_dates
```

```python
from datetime import timedelta, date

def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)
```

```python
by_date = pd.DataFrame(data_df['datetimes'].value_counts())

start_date = date(2000, 3, 5)
end_date = date(2014, 7, 23)

date_array = np.zeros(5253)
i = 0
for single_date in daterange(start_date, end_date):
    b = 1
    for idx, value in by_date.iterrows():
        if pd.to_datetime(single_date.strftime("%Y-%m-%d %H:%M:%S")) == idx:
            date_array[i] = value[0]
            b = 0
    if b == 1:
        date_array[i] = 0
    i = i + 1
```

```python
x = np.linspace(0,len(date_array),len(date_array))
dates = []
for single_date in daterange(start_date, end_date):
    dates.append(single_date)
plt.figure(figsize=(12,6))
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.title('Number of Reviews Per Day')
plt.plot(dates, date_array)
plt.show()
```

This visualization provides context surrounding the dates across which reviews were assembled for this dataset. Starting in around 2011, the volume of daily reviews begins to increase dramatically, and by 2013 the variation becomes quite significant in the region of 1200-1300 reviews per day. This indicates to us that further analysis of the dataset should begin in 2011 or even 2012 to ensure that the sample size is large enough to create a model of strong enough predictive power in both training/testing environments.

```python
dates_new = dates[-1000:]
date_array_new = date_array[-1000:]
plt.figure(figsize=(12,6))
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.title('Number of Reviews Per Day')
plt.plot(dates_new, date_array_new)
plt.show()
```

```python
averages = [np.mean(lengths1), np.mean(lengths2), np.mean(lengths3), np.mean(lengths4), np.mean(lengths5)]
x = np.linspace(1,5,5)

plt.bar(x, averages,color="lightskyblue")
plt.xlabel('Rating')
plt.ylabel('Average Word Count')
plt.title('Rating vs. Average Word Count')
plt.show()
```

This visualization aggregates the word counts for reviews that fall into the 5 categorical review ratings to determine whether there exists a correlation between a user's likeliness of recommending a product and how long their written review was. In general, what this visualization indicates that people who like their products more tend to write longer reviews for them, although this is not true for the highest categorical score of "5." The difference between the average word count length for category 4 ratings, however, is to the order of 40% longer. This reveals an important insight about the power of predicting a review's score based on its text, without even considering the text itself.

```python
a = pd.DataFrame(review_helpfulness.groupby(['proportions'])['lengths'].mean(), review_helpfulness.groupby(['proportions'])['lengths'].mean().index)
a = a.reset_index()
a.head()

plt.figure(figsize=(13,6))
plt.scatter(a['proportions'],a['lengths'],s=3,color="blue")
plt.xlabel('Proportion of Helpfulness')
plt.ylabel('Review Length')
plt.title('Review Length vs. Perceived Helpfulness')
plt.show()
```

This visualization provides information surrounding the relationship between the length of a given review and the degree to which the review was found helpful (calculated as the ratio of the number users who found the review helpful to the number of users who tagged the review as either helpful or unhelpful). Interestingly, although much quantitative analysis can follow from this observation, it seems that shorter reviews fall across the entire spectrum of helpfulness; that is, knowing that a review is short alone has very little ability for us to predict how helpful it will be. However, if a review is long, that is not necessarily the case: in the region beginning at around 4000 characters, the space of helpfulness is onstrained to a much higher region of the scale to the above 0.5 or above 0.6 threshold. This indicates that in genera, longer reviews are indeed more helpful.
